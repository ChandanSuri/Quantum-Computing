## Problem 1.1

Compute the single- and two-qubit reduced density matrices of the $3$-qubit GHZ state

$$
        \frac{\ket{000} + \ket{111}}{\sqrt{2}}~.
$$

## Solution

Density matrix for the GHZ state can be written as:
$$
\ket{GHZ}\bra{GHZ} = \left[\frac{1}{\sqrt{2}}(\ket{000} + \ket{111})\right]\left[\frac{1}{\sqrt{2}}(\bra{000} + \bra{111})\right]
$$
This can also be written as:
$$
\ket{GHZ}\bra{GHZ} = \frac{1}{2} \left( \ket{000}\bra{000} + \ket{111}\bra{000} + \ket{000}\bra{111} + \ket{111}\bra{111} \right )
$$

As there are 3 qubits here, to find the single and two-qubit reduced density matrices, I will have to find the trace here. Treating first qubit from different system and 2nd and 3rd qubits of a different one, we can find the traces as follows:

Trace w.r.t the 1st qubit (giving two-qubit reduced density matrix) can be written as:
$$
\rho_{23} = Tr_{1}(\ket{GHZ}\bra{GHZ}) = \ket{0}_{1}\ket{GHZ}\bra{GHZ}\bra{0}_{1} + \ket{1}_{1}\ket{GHZ}\bra{GHZ}\bra{1}_{1}
$$
This can further be written down as:
$$
\rho_{23} = Tr_{1}(\ket{GHZ}\bra{GHZ}) = \frac{1}{2} \left[ (|\langle 0 | 0 \rangle|^2 \ket{00}\bra{00} + \langle 0 | 1 \rangle \langle 0 | 0 \rangle \ket{11}\bra{00} + \langle 0 | 0 \rangle \langle 1 | 0 \rangle \ket{00}\bra{11} + \langle 0 | 1 \rangle \langle 1 | 0 \rangle \ket{11}\bra{11}) + (\langle 1 | 0 \rangle \langle 0 | 1 \rangle \ket{00}\bra{00} + \langle 1 | 1 \rangle \langle 0 | 1 \rangle \ket{11}\bra{00} + \langle 1 | 0 \rangle \langle 1 | 1 \rangle \ket{00}\bra{11} + |\langle 1 | 1 \rangle|^2 \ket{11}\bra{11}) \right]
$$
The values for the parts in the above equation where one part of ket-bra is 0 and another is 1, this gives us 0 which when multiplied to corresponding terms of a term will give out zero thus, making most of the terms in the above equation to be zero as follows:
$$
\rho_{23} = Tr_{1}(\ket{GHZ}\bra{GHZ}) = \frac{1}{2} \left[ (|\langle 0 | 0 \rangle|^2 \ket{00}\bra{00} + 0 + 0 + 0) + (0 + 0 + 0 + |\langle 1 | 1 \rangle|^2 \ket{11}\bra{11})\right]
$$
This gives us the following:
$$
\rho_{23} = Tr_{1}(\ket{GHZ}\bra{GHZ}) = \frac{1}{2} (|\langle 0 | 0 \rangle|^2 \ket{00}\bra{00} + |\langle 1 | 1 \rangle|^2 \ket{11}\bra{11})
$$
The values like $|\langle 0 | 0 \rangle|^2$ and $|\langle 1 | 1 \rangle|^2$ will become $1$, thus, giving us the following:
$$
\rho_{23} = Tr_{1}(\ket{GHZ}\bra{GHZ}) = \frac{1}{2} (\ket{00}\bra{00} + \ket{11}\bra{11})
$$

Next, Trace w.r.t the 2nd and 3rd qubit (giving one-qubit reduced density matrix) can be written as:
$$
\rho_{1} = Tr_{23}(\ket{GHZ}\bra{GHZ}) = \ket{00}_{23}\ket{GHZ}\bra{GHZ}\bra{00}_{23} + \ket{11}_{23}\ket{GHZ}\bra{GHZ}\bra{11}_{23}
$$
This can be further written down as:
$$
\rho_{1} = Tr_{23}(\ket{GHZ}\bra{GHZ}) =  \frac{1}{2} \left[ (|\langle 00 | 00 \rangle|^2 \ket{0}\bra{0} + \langle 00 | 11 \rangle \langle 00 | 00 \rangle \ket{1}\bra{0} + \langle 00 | 00 \rangle \langle 11 | 00 \rangle \ket{0}\bra{1} + \langle 00 | 11 \rangle \langle 11 | 00 \rangle \ket{1}\bra{1}) + (\langle 11 | 00 \rangle \langle 00 | 11 \rangle \ket{0}\bra{1} + \langle 11 | 11 \rangle \langle 00 | 11 \rangle \ket{1}\bra{0} + \langle 11 | 00 \rangle \langle 11 | 11 \rangle \ket{0}\bra{1} + |\langle 11 | 11 \rangle|^2 \ket{1}\bra{1}) \right]
$$
In the above equation, the values like $\langle 00 | 11 \rangle$ and $\langle 11 | 00 \rangle$ are actually zeros thus making the corresponding terms zero as well:
$$
\rho_{1} = Tr_{23}(\ket{GHZ}\bra{GHZ}) =  \frac{1}{2} \left[(|\langle 00 | 00 \rangle|^2 \ket{0}\bra{0} + 0 + 0 + 0) + (0 + 0 + 0 + |\langle 11 | 11 \rangle|^2 \ket{1}\bra{1})\right]
$$
Also, the values like $|\langle 00 | 00 \rangle|^2$ and $|\langle 11 | 11 \rangle|^2$ will become $1$, thus, giving us the following:
$$
\rho_{1} = Tr_{23}(\ket{GHZ}\bra{GHZ}) =  \frac{1}{2} \left[\ket{0}\bra{0} + \ket{1}\bra{1}\right]
$$

So, now that we have both the traces, essentially, whole density matrix can be written as a combination of these 2 reduced density matrices as:
$$
 = \rho_{1} \otimes \rho_{23}
$$
This, in turn can be written as:
$$
 = \frac{1}{4} (\ket{0}\bra{0} + \ket{1}\bra{1}) \otimes (\ket{00}\bra{00} + \ket{11}\bra{11})
$$

## Problem 1.2

Let $\{ (p_1,\ket{\psi_1}),\ldots,(\psi_k,\ket{\psi_k}) \}$ denote a probabilistic mixture of states -- not necessarily orthogonal -- where $p_i$'s are probabilities that sum to $1$. Consider the density matrix $\sigma = p_1 \ketbra{\psi_1}{\psi_1} + \cdots + p_k \ketbra{\psi_k}{\psi_k}$ on a system $A$. Show that the following vector $\ket{\theta}_{AB}$ is a valid pure state (i.e. unit length vector), and it purifies $\sigma$:

$$
        \ket{\theta} = \sum_{i = 1}^k \sqrt{p_i} \ket{\psi_i}_A \otimes \ket{i}_B
$$

where $B$ is a register of dimension at least $k$ and $\{ \ket{i} \}$ is an orthonormal basis for $B$.

## Solution

We know that if the vector $\ket{\theta}_{AB}$ is a valid pure state, then the following should be proven true:
$$
\sigma_A = Tr_B (\ket{\theta}_{AB} \bra{\theta}_{AB})
$$

Let's try and prove this then!

We put the formulation for $\ket{\theta}_{AB}$ in the above formulation as follows:
$$
\sigma_A = Tr_B \left( \sum_{i=1}^k \sqrt{p_i} \ket{\psi_i}_A \otimes \ket{i}_B \cdot \sum_{i=1}^k \sqrt{p_i} \bra{i}_B \otimes \bra{\psi_i}_A \right)
$$

As, $\{ \ket{i} \}$ is an orthonormal basis for $B$, we know that the vectors in the basis when multiplied with any vector in the basis but itself results in a zero, so, we won't take those terms into consideration and thus, we can join the two summations as one as follows:
$$
\sigma_A = Tr_B \left( \sum_{i=1}^k p_i (\ket{\psi_i}_A \otimes \ket{i}_B) \cdot (\bra{i}_B \otimes \bra{\psi_i}_A) \right)
$$

As this is a trace for the system B, we can take out the terms associated with system A outside of the Trace operation giving us the following:
$$
\sigma_A = \sum_{i=1}^k p_i \ket{\psi_i}_A \bra{\psi_i}_A \left[ Tr_B \left( \sum_{i=1}^k \ket{i}_B \bra{i}_B \right) \right]
$$

Since, trace is linear, we can move the trace operator inside the summation as follows:
$$
\sigma_A = \sum_{i=1}^k p_i \ket{\psi_i}_A \bra{\psi_i}_A \left[ \sum_{i=1}^k Tr_B ( \ket{i}_B \bra{i}_B )\right]
$$
Using the property of trace where $Tr (\ket{\psi} \bra{\psi}) = \langle \psi | \psi \rangle$, we can write the above formula as:
$$
\sigma_A = \sum_{i=1}^k p_i \ket{\psi_i}_A \bra{\psi_i}_A \left[ \sum_{i=1}^k \langle i | i \rangle \right]
$$

Next, since $\{ \ket{i} \}$ is an orthonormal basis for $B$, multiplying the vectors in the basis with itself and adding them up will give us $1$ over all the dimensions. Thus, the formulation becomes as follows:
$$
\sigma_A = \sum_{i=1}^k p_i \ket{\psi_i}_A \bra{\psi_i}_A \left[ 1 \right]
$$

Thus, this becomes:
$$
\sigma_A = \sum_{i=1}^k p_i \ket{\psi_i}_A \bra{\psi_i}_A
$$

This is basically the same as LHS as:
$$
\sigma_A = p_1 \ket{\psi_1}_A \bra{\psi_1}_A + p_2 \ket{\psi_2}_A \bra{\psi_2}_A + ... + p_k \ket{\psi_k}_A \bra{\psi_k}_A
$$

Hence, we proved that the vector $\ket{\theta}_{AB}$ is a valid pure state (i.e. unit length vector), and it purifies $\sigma$.

\ket{\psi}_{AB} = \sqrt{\frac{1}{2}} \left(\left(\begin{array}{c}

## Problem 1.4

Let $\ket{\psi} = \alpha \ket{0} + \beta \ket{1}$ denote an unknown qubit, and suppose it was just teleported from Alice to Bob. But imagine that, before Alice is able to tell Bob the outcomes of her measurements, her internet cuts out. So Bob's qubit is one of the states

$$
    \ket{\psi}, X \ket{\psi}, Z \ket{\psi}, XZ \ket{\psi}
$$

with equal probability. What is the density matrix that describes the mixed state of Bob's qubit?

## Solution

As $\ket{\psi} = \alpha \ket{0} + \beta \ket{1}$, therefore:
$$
X \ket{\psi} = \beta \ket{0} + \alpha \ket{1},
$$
$$
Z \ket{\psi} = \alpha \ket{0} - \beta \ket{1},
$$
$$
XZ \ket{\psi} = - \beta \ket{0} + \alpha \ket{1}
$$

The density matrix for these can be written as (considering that the probabilities for each of these is equal and equal to $\frac{1}{4}$):
$$
\rho = \frac{1}{4} \left( (\alpha \ket{0} + \beta \ket{1})(\alpha \bra{0} + \beta \bra{1}) + (\beta \ket{0} + \alpha \ket{1})(\beta \bra{0} + \alpha \bra{1}) + (\alpha \ket{0} - \beta \ket{1})(\alpha \bra{0} - \beta \bra{1}) + (- \beta \ket{0} + \alpha \ket{1})(- \beta \bra{0} + \alpha \bra{1})\right)
$$

When opened up as matrices, this looks like:
$$
\rho = \frac{1}{4} \left( \left(\begin{array}{c}
\alpha \\
\beta
\end{array}\right) \left(\begin{array}{cc}
\alpha & \beta
\end{array}\right) + \left(\begin{array}{c}
\beta \\
\alpha
\end{array}\right) \left(\begin{array}{cc}
\beta & \alpha
\end{array}\right) + \left(\begin{array}{c}
\alpha \\
-\beta
\end{array}\right) \left(\begin{array}{cc}
\alpha & -\beta
\end{array}\right) + \left(\begin{array}{c}
-\beta \\
\alpha
\end{array}\right) \left(\begin{array}{cc}
-\beta & \alpha
\end{array}\right) \right)
$$

Solving it gives the following:
$$
\rho = \frac{1}{4} \left( \left(\begin{array}{cc}
\alpha^2 & \alpha \beta \\
\alpha \beta & \beta^2
\end{array}\right) + \left(\begin{array}{cc}
\beta^2 & \alpha \beta \\
\alpha \beta & \alpha^2
\end{array}\right) + \left(\begin{array}{cc}
\alpha^2 & -\alpha \beta \\
-\alpha \beta & \beta^2
\end{array}\right) + \left(\begin{array}{cc}
\beta^2 & -\alpha \beta \\
-\alpha \beta & \alpha^2
\end{array}\right)\right)
$$

Solving it further gives us:
$$
\rho = \frac{1}{4} \left(\begin{array}{cc}
2(\alpha^2 + \beta^2) & 0 \\
0 & 2(\alpha^2 + \beta^2)
\end{array}\right) = \frac{1}{2} \left(\begin{array}{cc}
\alpha^2 + \beta^2 & 0 \\
0 & \alpha^2 + \beta^2
\end{array}\right)
$$

This is the final density matrix above.

# Problem 2: Limitations of Shor's code

Show that there are distinct two-qubit errors $E_1 \neq E_2$ (unitaries acting on two qubits) such that

$$
    E_1 \ket{\overline{0}} = E_2 \ket{\overline{1}}
$$

where $\ket{\overline{0}},\ket{\overline{1}}$ represent the logical $\ket{0},\ket{1}$ state using the $9$-qubit Shor code. This shows that the Shor code is not capable of correcting more than $1$ error.

## Solution

Initially the Shor's code for $\ket{\overline{0}}$ and $\ket{\overline{1}}$ can be written as:
\begin{gather*}
    \ket{\overline{0}} = \frac{1}{\sqrt{8}} \Big( \ket{000000000} + \ket{000111000} + \ket{111000000} + \ket{111111000} + \ket{000000111} + \ket{000111111} + \ket{111000111} + \ket{111111111} \Big)  \\
    \ket{\overline{1}} = \frac{1}{\sqrt{8}} \Big( \ket{000000000} - \ket{000111000} - \ket{111000000} + \ket{111111000} - \ket{000000111} + \ket{000111111} + \ket{111000111} - \ket{111111111} \Big)
\end{gather*}

Let's take the following distinct two-qubit errors:
\begin{gather*}
    E_1 = X_1 Z_7 \\
    E_2 = X_1 Z_5 Z_1
\end{gather*}

First, let's compute $E_1 \ket{\overline{0}} = X_1 Z_7 \ket{\overline{0}}$:
$$
E_1 \ket{\overline{0}} = X_1 Z_7 \ket{\overline{0}} = X_1 \Big( \frac{1}{\sqrt{8}} \Big( \ket{000000000} + \ket{000111000} + \ket{111000000} + \ket{111111000} - \ket{000000111} - \ket{000111111} - \ket{111000111} - \ket{111111111} \Big) \Big)
$$
Solving further:
$$
E_1 \ket{\overline{0}} = X_1 Z_7 \ket{\overline{0}} = \frac{1}{\sqrt{8}} \Big( \ket{100000000} + \ket{100111000} + \ket{011000000} + \ket{011111000} - \ket{100000111} - \ket{100111111} - \ket{011000111} - \ket{011111111} \Big)
$$

Second, let's compute $E_2 \ket{\overline{1}} = X_1 Z_5 Z_1 \ket{\overline{1}}$:
$$
E_2 \ket{\overline{1}} = X_1 Z_5 Z_1 \ket{\overline{1}} = X_1 Z_5 \Big( \frac{1}{\sqrt{8}} \Big( \ket{000000000} - \ket{000111000} + \ket{111000000} - \ket{111111000} - \ket{000000111} + \ket{000111111} - \ket{111000111} + \ket{111111111} \Big) \Big)
$$
Solving further:
$$
E_2 \ket{\overline{1}} = X_1 Z_5 Z_1 \ket{\overline{1}} = X_1 \Big( \frac{1}{\sqrt{8}} \Big( \ket{000000000} + \ket{000111000} + \ket{111000000} + \ket{111111000} - \ket{000000111} - \ket{000111111} - \ket{111000111} - \ket{111111111} \Big) \Big)
$$
Solving even further:
$$
E_2 \ket{\overline{1}} = X_1 Z_5 Z_1 \ket{\overline{1}} = \frac{1}{\sqrt{8}} \Big( \ket{100000000} + \ket{100111000} + \ket{011000000} + \ket{011111000} - \ket{100000111} - \ket{100111111} - \ket{011000111} - \ket{011111111} \Big)
$$

As we can see above, using two distinct two-qubit errors gives us the same result such that:
$$
E_1 \ket{\overline{0}} = E_2 \ket{\overline{1}}
$$

As such a pair exists, thus, this shows that the Shor code is not capable of correcting more than $1$ error.

# Problem 3: An error _detection_ code

Shor's $9$-qubit code can **correct** any single-qubit error on one of its $9$ qubits. Below we give a $4$-qubit code which can \emph{detect} any single-qubit error on its qubits. By this we mean that there is a detection circuit that determines if a single-qubit error has occurred, but it can't necessarily identify which one has occurred. 

The encoding map is as follows:
\begin{gather*}
    \ket{\overline{0}} = \frac{1}{2} \Big( \ket{00} + \ket{11} \Big) \otimes \Big( \ket{00} + \ket{11} \Big) \\
    \ket{\overline{1}} = \frac{1}{2} \Big( \ket{00} - \ket{11} \Big) \otimes \Big( \ket{00} - \ket{11} \Big)~.
\end{gather*}

## Problem 3.1

Give an example of two distinct single-qubit unitaries $E_1 \neq E_2$ (which may act on different qubits) such that

$$
    E_1 \ket{\overline{0}} = E_2 \ket{\overline{1}}~.
$$

This shows that this code cannot uniquely identify single-qubit errors, because given $E_1 \ket{\overline{0}}$, one can't be sure whether the original state was $\ket{\overline{0}}$ or $\ket{\overline{1}}$. However, in the next few parts you will show that the code can still detect that **some** error has occurred. 

## Solution

The equations are specified as:
\begin{gather*}
    \ket{\overline{0}} = \frac{1}{2} \Big( \ket{00} + \ket{11} \Big) \otimes \Big( \ket{00} + \ket{11} \Big) \\
    \ket{\overline{1}} = \frac{1}{2} \Big( \ket{00} - \ket{11} \Big) \otimes \Big( \ket{00} - \ket{11} \Big)~.
\end{gather*}

If we take $E_1 = Z_1$ and $E_2 = Z_3$, that means that $E_1$ is the single qubit unitary for applying Z gate on $1^{st}$ qubit and $E_2$ is the single qubit unitary for applying Z gate on $3^{rd}$ qubit as follows:
\begin{gather*}
    E_1 \ket{\overline{0}} = Z_1 \ket{\overline{0}} = \frac{1}{2} \Big( \ket{00} - \ket{11} \Big) \otimes \Big( \ket{00} + \ket{11} \Big) \\
    E_2 \ket{\overline{1}} = Z_3 \ket{\overline{1}} = \frac{1}{2} \Big( \ket{00} - \ket{11} \Big) \otimes \Big( \ket{00} + \ket{11} \Big)~.
\end{gather*}

As we can see above, applying $E_1 = Z_1$ and $E_2 = Z_3$ above gives us $E_1 \ket{\overline{0}} = E_2 \ket{\overline{1}}~$.

As those were two different single qubit unitaries acting on different qubits resulting in the same state, these two distinct single qubit unitaries show that this code cannot uniquely identify single-qubit errors, because given $E_1 \ket{\overline{0}}$, one can't be sure whether the original state was $\ket{\overline{0}}$ or $\ket{\overline{1}}$.

## Problem 3.2

Give a procedure (either as a quantum circuit or sufficiently detailed pseudocode) that detects a \emph{bitflip} error on a single qubit of an encoded state $\alpha \ket{\overline{0}} + \beta \ket{\overline{1}}$. In other words the procedure, if it's given a valid encoded state $\ket{\overline{\psi}}$, returns $\ket{\overline{\psi}}$ and outputs "No error", whereas if it's given $X_i \ket{\overline{\psi}}$ where $X_i$ is the bitflip operation on some qubit $i$, then the circuit outputs "$X$ error somewhere". 

## Solution

The encoded state can be written as follows:
$$
\ket{\overline{\psi}} = \alpha \ket{\overline{0}} + \beta \ket{\overline{1}} = \frac{\alpha}{2} \Big( \ket{00} + \ket{11} \Big) \otimes \Big( \ket{00} + \ket{11} \Big) + \frac{\beta}{2} \Big( \ket{00} - \ket{11} \Big) \otimes \Big( \ket{00} - \ket{11} \Big)
$$

Solving it further gives us the following:
$$
\ket{\overline{\psi}} = \alpha \ket{\overline{0}} + \beta \ket{\overline{1}} = \frac{\alpha}{2} \Big( \ket{0000} + \ket{1100} + \ket{0011} + \ket{1111} \Big) + \frac{\beta}{2} \Big( \ket{0000} - \ket{0011} - \ket{1100} + \ket{1111} \Big)
$$

Now, let's say there was a bit flip error on the 2nd qubit $X_2 \ket{\overline{\psi}}$:
$$
X_2 \ket{\overline{\psi}} = \frac{\alpha}{2} \Big( \ket{0100} + \ket{1000} + \ket{0111} + \ket{1011} \Big) + \frac{\beta}{2} \Big( \ket{0100} - \ket{0111} - \ket{1000} + \ket{1011} \Big)
$$

If we use one syndrome qubits (as explained below), the complete state of the system can be defined as:
$$
\frac{\alpha}{2} \Big( \ket{0100} + \ket{1000} + \ket{0111} + \ket{1011} \Big) \ket{0} + \frac{\beta}{2} \Big( \ket{0100} - \ket{0111} - \ket{1000} + \ket{1011} \Big) \ket{0}
$$

Coming to the procedure we could use, we can do the following:
<ol>
    <li>We will take 1 syndrome qubit here that we will be measuring at the end to see if any error might have occured.
    </li>
    <li>We will put a CNOT gate between the 1st qubit and the syndrome qubit such that the control qubit is the 1st qubit of the encoded state and target qubit is the syndrome qubit. Also, we put another CNOT after that between the 2nd qubit of the encoded state and the syndrome qubit such that the control qubit is from the encoded state and the target qubit is the syndrome qubit.
    </li>
    <li>In a similar fashion as above, we put another CNOT gate between the 3rd qubit of the encoded state and the syndrome qubit, and between the 4th qubit of the encoded state and the syndrome qubit (the ones from the encoded state are control qubits and the syndrome qubit is the target qubit).
    </li>
    <li>Then, at the end we perform the measurements on the syndrome qubit. If, the syndrome qubit is in $\ket{0}$ state then we can say with certainity that there is no bit flip error. But if the syndrome qubit is in $\ket{1}$ state, then we can say that there is some error although it doesn't certainly tell us the which qubit there is an error in. As, we are creating a bit flip error detection circuit, using this procedure will work.
    </li>
</ol>

\frac{\alpha}{2} \Big( \ket{0100} + \ket{1000} + \ket{0111} + \ket{1011} \Big) \ket{1} + \frac{\beta}{2} \Big( \ket{0100} - \ket{0111} - \ket{1000} + \ket{1011} \Big) \ket{1}

$$
\frac{\alpha}{2} \Big( \ket{0100} \ket{1} + \ket{1000} \ket{1} + \ket{0111} \ket{0} + \ket{1011} \ket{0} \Big) + \frac{\beta}{2} \Big( \ket{0100} \ket{1} - \ket{0111} \ket{0} - \ket{1000} \ket{1} + \ket{1011} \ket{0} \Big)

As we can see above, we are getting $\ket{1}$ state at the end because there was a bit flip error. If we look very closely, we can see that if there is any single bit flip error, then there will be unequal number of 0's and 1's because of which the syndrome qubit will end up as $\ket{1}$. Precisely, if there is a bitflip error on any one of the qubits, then the $\ket{\overline{\psi}}$ will contains odd number of qubit in state 1, thus, applying CNOT odd number of times on the syndrome qubit will result in state $\ket{1}$.  Thus, this procedure when gives $\ket{0}$ as an output after measurement then that means that "No error". Otherwise, we can say that "X error somewhere".

## Problem 3.4

Explain how to detect **any** unitary $1$-qubit error on one of the $4$ qubits.

## Solution

To detect any unitary 1-qubit error one any one of the 4 qubits, we know there can be only one of the four following cases (considering C as the circuit explained below, $s_1$ syndrome qubit needed for bitflip error and $s_2$ for phaseflip error):
<ol>
    <li> Only Bitflip error: $C X_i \ket{\overline{\psi}} \ket{00} = X_i \ket{\overline{\psi}} \ket{10}$
    </li>
    <li> Only Phaseflip error: $C Z_i \ket{\overline{\psi}} \ket{00} = Z_i \ket{\overline{\psi}} \ket{01}$
    </li>
    <li> Both Bitflip and Phaseflip error: $C Y_i \ket{\overline{\psi}} \ket{00} = Y_i \ket{\overline{\psi}} \ket{11}$
    </li>
    <li> No error: $C I \ket{\overline{\psi}} \ket{00} = \ket{\overline{\psi}} \ket{00}$
    </li>
</ol>

For the 1st case above, we can use the procedure or circuit shown and explained in problem 3.2 solution.

For the 2nd case above, we can use the procedure or circuit shown and explained in problem 3.3 solution.

For the 3rd case above, we can combine both the procedures for bit flip detection and phaseflip detection one after the other and use separate syndrome qubits for each of them (1 for phaseflip detection and 1 for bitflip detection). 

For the 4th case, whatever we do there is no error to detect!

Thus,combining both the procedures one after the other would detect any unitary 1-qubit error on any one of the 4 qubits successfully!

Moreover, any unitary $M$ can be written as a linear combination of the Pauli matrices as $M = aI + bX + cZ + dY$. Then, applying the circuit $C$ to $M \ket{\overline{\psi}}$, we would get the following:
$$
C M \ket{\overline{\psi}} \ket{00} = a \times C I \ket{\overline{\psi}} \ket{00} + b \times C X_i \ket{\overline{\psi}} \ket{00} + c \times C Z_i \ket{\overline{\psi}} \ket{00} + d \times C Y_i \ket{\overline{\psi}} \ket{00}
$$

Using the formulation above, we can also write this as:
$$
C M \ket{\overline{\psi}} \ket{00} = a \times \ket{\overline{\psi}} \ket{00} + b \times X_i \ket{\overline{\psi}} \ket{10} + c \times Z_i \ket{\overline{\psi}} \ket{01} + d \times Y_i \ket{\overline{\psi}} \ket{11}
$$

Now, based on the measurement we make on the syndrome qubits, we would get one of the states out of the 4 possible cases. Thus, if we measure $00$ on the syndrome qubits, that would mean that there was no error. Also, $10$ would mean the presence of bit flip error, $01$ would mean the presence of phase flip error, and $11$ would mean the presence of both bit flip and phase flip error.

## Problem 4.1 - Benchmarking individual qubits



In this subproblem, we will benchmark individual qubits on (a simulation of) the 5-qubit `ibmq_essex` device. A typical way to benchmark a qubit is to run a sequence of randomly chosen single-qubit gates $g_1,g_2,\ldots,g_k$, and then running the reverse sequence $g_k^{-1},g_{k-1}^{-1},\ldots,g_1^{-1}$ so that overall the effect should be the identity. Of course, each gate will incur some noise, so the state of the qubit will drift over time. One can measure the noise by measuring the qubit at the end of the sequence to see if it stayed in the $\ket{0}$ state. 

You will write code to perform the following: for $k = 10, 20, 30, \ldots,100$, for each qubit $q = 0, 1,2, \ldots,4$, pick a sequence of gates $g_{1},\ldots,g_k$ where each $g_i$ is chosen randomly from the gate set $\{ X, Y,Z, H \}$. Then, on qubit $q$ run the sequence $(g_i)$ forward and then in reverse, and measure the qubit. Do this 1000 times and calculate the percentage $p_{q,k}$ of times that the qubit $q$ ends back in the $\ket{0}$ state. 

Below, we've provided two functions. The first function, `benchmark_qubit`, requires you to fill in some code, and it returns a `IBMQJob` object. The second function, `retrieve_job_results`, takes an `IBMQJob` object and gets the measurement counts. (The reason we're dividing this up into two steps is because later we'll run these circuits on a real device).

```python
from qiskit import IBMQ, transpile
from qiskit import QuantumCircuit
from qiskit.providers.aer import AerSimulator
from qiskit.providers.fake_provider import FakeEssex
from qiskit.providers.jobstatus import JobStatus

import numpy as np
from matplotlib import pyplot as plt
import random

##########
# This function takes as input
#    - q : qubit number 0 thru 4
#    - k : length of random gate sequence
#    - shots: number of times to run and measure the sequence
#    - device: the device (either simulated or real) to run the circuit on
#
#  Returns: 
#    - the IBMQJob object corresponding to the circuit (see https://qiskit.org/documentation/stubs/qiskit.providers.ibmq.job.IBMQJob.html#qiskit.providers.ibmq.job.IBMQJob)
##########
def benchmark_qubit(q, k, shots, device):
    
    circ = QuantumCircuit(5, 1)

    ### WRITE CODE TO GENERATE THE RANDOM SEQUENCE OF LENGTH k, and ITS REVERSAL, ON QUBIT q ####################
    gate_set = set(['X', 'Y', 'Z', 'H'])
    gates_selected = list()
    
    for curr_gate_idx in range(k):
        curr_gate = random.sample(gate_set, 1)[0]
        gates_selected.append(curr_gate)
        if curr_gate == 'X':
            circ.x(q)
        elif curr_gate == 'Y':
            circ.y(q)
        elif curr_gate == 'Z':
            circ.z(q)
        elif curr_gate == 'H':
            circ.h(q)
        else:
            raise Exception("The gate selected is not valid!")
    
    for curr_gate_idx in range(k-1, -1, -1):
        curr_gate = gates_selected[curr_gate_idx]
        if curr_gate == 'X':
            circ.x(q)
        elif curr_gate == 'Y':
            circ.y(q)
        elif curr_gate == 'Z':
            circ.z(q)
        elif curr_gate == 'H':
            circ.h(q)
        else:
            raise Exception("The gate selected is not valid!")
        
    ### END CODE BLOCK ###################
    
    # measure qubit q, and store it in classical register [0]
    circ.measure([q], [0])
    
    #this will break down the circuit into gates that can be implemented on the chip
    compiled_circuit = transpile(circ,device,optimization_level=0)
    
    job = device.run(compiled_circuit,shots=shots)
    return job

##########
# This function takes as input
#    - job: the IBMQJob object corresponding to a circuit being executed
#    - blocking: if True, then this will wait until the circuit results are done executing on the device
#                (either fake or real). If False, then this will first check the status of the job.
#
#  Returns: 
#    - if the job is done, then it returns the counts as a dictionary (e.g., { '0000': 356, '0001': 288, ...})
#      otherwise if the job is still running or some other status, then it returns None
##########
def retrieve_job_results(job,blocking=True):
    
    #if it's blocking, then just go ahead and call result()
    if blocking:
        counts = job.result().get_counts(0)
        return counts
    else:
        #first, check the status
        job_status = job.status()
    
        if job_status is JobStatus.DONE:
            counts = job.result().get_counts(0)
            return counts            
        else:
            print("The job ",job.job_id," has status: ",job_status)
            return None

        
```

```python
##########
# This function takes as input
#    - r : size of GHZ state
#    - k : length of random gate sequence
#    - shots: number of times to run and measure the sequence
#    - device: the device to run on, either simulated or real
#
#  Returns: 
#    - the IBMQJob object corresponding to the circuit (see https://qiskit.org/documentation/stubs/qiskit.providers.ibmq.job.IBMQJob.html#qiskit.providers.ibmq.job.IBMQJob)
##########
def benchmark_circuit(r, k, shots, device):
    
    circ = QuantumCircuit(5, r)

    ### WRITE CODE TO GENERATE C_r and its reversal k times ####################
    
             ## You need to choose your subset of qubits carefully!
    if r == 2:
        qubits_to_use = [1, 2]
        for curr_k in range(k):
            circ.h(qubits_to_use[0])
            circ.cx(qubits_to_use[1], qubits_to_use[0])
            circ.cx(qubits_to_use[1], qubits_to_use[0])
            circ.h(qubits_to_use[0])
    elif r == 3:
        qubits_to_use = [2, 1, 3]
        for curr_k in range(k):
            circ.h(qubits_to_use[0])
            circ.cx(qubits_to_use[1], qubits_to_use[0])
            circ.cx(qubits_to_use[2], qubits_to_use[1])
            circ.cx(qubits_to_use[2], qubits_to_use[1])
            circ.cx(qubits_to_use[1], qubits_to_use[0])
            circ.h(qubits_to_use[0])
    elif r == 4:
        qubits_to_use = [2, 1, 3, 0]
        for curr_k in range(k):
            circ.h(qubits_to_use[0])
            circ.cx(qubits_to_use[1], qubits_to_use[0])
            circ.cx(qubits_to_use[2], qubits_to_use[1])
            circ.cx(qubits_to_use[3], qubits_to_use[1])
            circ.cx(qubits_to_use[3], qubits_to_use[1])
            circ.cx(qubits_to_use[2], qubits_to_use[1])
            circ.cx(qubits_to_use[1], qubits_to_use[0])
            circ.h(qubits_to_use[0])
    elif r == 5:
        qubits_to_use = [2, 1, 3, 0, 4]
        for curr_k in range(k):
            circ.h(qubits_to_use[0])
            circ.cx(qubits_to_use[1], qubits_to_use[0])
            circ.cx(qubits_to_use[2], qubits_to_use[1])
            circ.cx(qubits_to_use[3], qubits_to_use[1])
            circ.cx(qubits_to_use[4], qubits_to_use[2])
            circ.cx(qubits_to_use[4], qubits_to_use[2])
            circ.cx(qubits_to_use[3], qubits_to_use[1])
            circ.cx(qubits_to_use[2], qubits_to_use[1])
            circ.cx(qubits_to_use[1], qubits_to_use[0])
            circ.h(qubits_to_use[0])
    else:
        raise Exception("This value of r is not valid!")

    ### END CODE BLOCK ###################
    
    
    ### WRITE CODE TO MEASURE AND COMPUTE P_RK #####################
    
    # measure r of the qubits , and store it in the r classical registers
    #for example, if r = 3, circ.measure([1,3,4],[0,1,2]) would mean measuring qubits 1,3,4
    
    # <--- WRITE MEASUREMENT CODE HERE
    circ.measure(qubits_to_use, list(range(r)))
    
    #this will break down the circuit into gates that can be implemented on the chip
    compiled_circuit = transpile(circ,device,optimization_level=0)
    job = device.run(compiled_circuit,shots=shots)
    
    return job

#### WRITE CODE TO CALL benchmark_circuit, retrieve_job_results AND PLOT p_rk versus k #################    

fake_device = AerSimulator.from_backend(FakeEssex())

#### WRITE YOUR CODE HERE #################    
from collections import defaultdict

prk_vals_all = defaultdict(dict)
for r in range(2, 6):
    prk_vals = dict()
    for k in range(10, 101, 10):
        job = benchmark_circuit(r,k,1000,fake_device)
        counts = retrieve_job_results(job)
        print(f'For r: {r} and k: {k}, the counts are as follows:{counts}')
        total_shots = sum(counts.values())
        prk_vals[k] = float(counts['0'*r])/total_shots
    
    prk_vals_all[r] = prk_vals
    
    # Plotting the graph
    plt.plot(prk_vals.keys(), prk_vals.values(), label = f'r={r}')

plt.xlabel('$k$')
plt.ylabel('$p_rk values$')
plt.title('Probability of getting ket <0|^r vs. k')
plt.legend()

plt.show()

#### END CODE BLOCK ############################################################
```

Below we provide some helper functions to save jobs to a file.

```python
##########
# This function takes as input
#    - list_of_jobs: self-explanatory
#    - filename: file to store the job_ids
##########
def save_jobs_to_file(list_of_jobs,filename):
    
    #open the file
    with open(filename,'w') as f:
        for job in list_of_jobs:
            f.write(job.job_id())
            f.write('\n')

##########
# This function takes as input
#    - name_of_device: this is the name of the device you submitted the jobs to, such as `ibmq_belem` or `ibm_nairobi`, etc.
#    - filename: file that has all the job_ids
#
# Returns: a list_of_jobs
##########            
def load_jobs_from_file(name_of_device,filename):
    IBMQ.load_account()
    provider = IBMQ.get_provider(hub='ibm-q-education')
    device = provider.get_backend(name_of_device)
    
    list_of_jobs = []
    with open(filename,'r') as f:
        job_ids = f.readlines()
        for job_id in job_ids:
            job = device.retrieve_job(job_id.strip())
            list_of_jobs.append(job)
            
    return list_of_jobs


##########
# This function takes as input
#    - list_of_jobs: self-explanatory
##########
def print_job_statuses(list_of_jobs):
    for job in list_of_jobs:
        print("id: ",job.job_id()," status: ", job.status())
        
```
